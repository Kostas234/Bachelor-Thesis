```python
# LCP(word by word)
 
# lcp for 2 strings
def lcp(x,y):

    result = ""
    n=min(len(x),len(y))
    
    for j in range(n):
        if x[j]!=y[j]:
            break
        result += x[j]
         
    return result
 
def LCP1(x):
     
    n=len(x)
    result=x[0]
    for i in xrange(1,n):
        result=lcp(result,x[i])
         
    return result           
```

```python
# LCP(character by character)
 
def LCP2(x):
   
	result=''
	n=len(x)
	m=len(x[0])
	for i in range(n-1):
		m=min(m,len(x[i+1]))
 
	for j in range(m):
		y=x[0][j]
		for k in range(n): 
			if x[k][j]!=y:
                		break
		else:
			continue
		break

	result += x[0][0:j+1] 

	return result
```

```python
# LCP divide and conquer

def lcp2(x,y):
    
    result = ""
    n=min(len(x),len(y))

    for j in range(n):
        if x[j]!=y[j]:
            break
        result += x[j]
        
    return result

def LCP(x,low,high):
    
    if low==high:
        return x[low]
    
    if high>low:
        mid=low+(high-low)/2
        
        str1=LCP(x,low,mid)
        str2=LCP(x,mid+1,high)
        
        result=lcp2(str1,str2)
        
        return result

    
```

```python
def LCP3(x):
    
    n=len(x)    
    prefix=""
    m=len(x[0])
    for i in range(n-1):
        m=min(m,len(x[i+1]))
        
    low=0
    high=m
    mid=low+(high-low)/2
    
    for i in range(1,n):
        for j in range(mid):
            if x[i][j]!=x[0][j]:
                prefix=x[0][0:j]
                break
        else:
            prefix=x[0][0:mid]
            
            for w in range(1,n):
                for v in range(mid,high):
                    if x[w][v]!=x[0][v]:
                        prefix=x[0][0:v]
                        break
                else:
                    prefix=x[0][0:high]
                    continue
                break
            continue
        break
        
    return prefix
```

```python
def longestCommonPrefix(a):
    
    def matching(first_str, a, start, end):
        for i in range(1, len(a)):
            if len(a[i][start:end+1]) < len(first_str):
                return
            if a[i][start:end+1] != first_str:
                return
        return first_str
    if not a:
        return ''
    sub_str = []
    low = 0
    high = len(a[0]) -1

    while(low <= high):
        mid = int(low + (high-low)/2)
        if matching(a[0][low:mid+1], a, low, mid):
            sub_str += a[0][low:mid+1]
            low = mid + 1
        else:
            high = mid - 1

    return(''.join(sub_str))
```
